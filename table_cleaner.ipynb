# Список дивизионов

```python
from bs4 import BeautifulSoup
from selenium import webdriver

# Отправить GET-запрос на страницу
url = "https://rosatom.ru/about/factories/"  # Замените на нужный URL
chrome_path = ""
options = webdriver.ChromeOptions()
options.add_argument("headless")
options.add_argument("--log-level=3")
options.add_experimental_option("excludeSwitches", ["enable-logging"])
browser = webdriver.Chrome(options=options)
# Создать объект BeautifulSoup для обработки содержимого страницы
browser.get(url)
html = browser.page_source
soup = BeautifulSoup(html, "html.parser")

# Найти все элементы h3 и получить их текст
h3_elements = soup.find_all("h3")

companies = [element.get_text().strip() for element in h3_elements]

```

```python
def string_match(string1, string2):
    matcher = SequenceMatcher(None, string1, string2)
    similarity = matcher.ratio()  # Получить коэффициент сходства (от 0 до 1)
    return similarity


def find_company(s):
    highest_sim = 0
    highest_div = ""
    for div in companies:
        sim = string_match(s, div)
        if sim > highest_sim:
            highest_sim = sim
            highest_div = div
    return highest_div
```

## Подбор дивизиона

```python
# Отправить GET-запрос на страницу
url = 'https://web.archive.org/web/20210928062830/https://rosatom-career.ru/center/companies-of-rosatom#division_1'  # Замените на нужный URL
chrome_path = ""
options = webdriver.ChromeOptions()
options.add_argument('headless')
options.add_argument("--log-level=3")
options.add_experimental_option('excludeSwitches', ['enable-logging'])
browser = webdriver.Chrome(options=options)
# Создать объект BeautifulSoup для обработки содержимого страницы
browser.get(url)
html = browser.page_source
soup = BeautifulSoup(html, 'html.parser')

divisions = dict()

for div_html in soup.find_all('div', {'class': 'division_companies'}):
    h3 = div_html.find('h3')
    if h3:
        division_name = h3.text.strip()
        # print(division_name)
        divisions[division_name] = []
        for company_html in div_html.find_all('div', {'class': 'division_companies--name'}):
            company_name = company_html.text.replace('"', '').strip()
            divisions[division_name].append(company_name)
```

```python
def find_division(s):
    highest_sim = 0
    highest_div = ""
    
    for div, companies in divisions.items():
        for company in companies:
            sim = string_match(s, company)
            
            if sim > highest_sim:
                highest_sim = sim
                highest_div = div
    return highest_div
```

```python
df["Дивизион"] = df["Предприятие"].apply(lambda x: find_division(str(x)))
```

```python
def format_date(date):
    if pd.isnull(date):  # Check for missing values
        return ''
    else:
        try:
            # Try to parse the date in "YYYY-MM-DD" format
            dt = datetime.strptime(date, '%Y-%m-%d')
        except ValueError:
            try:
                # Try to parse the date in "DD.MM.YYYY" format
                dt = datetime.strptime(date, '%d.%m.%Y')
            except ValueError:
                try:
                    # Try to parse the date in "месяц.год" format
                    dt = datetime.strptime(date, '%b.%Y')
                except ValueError:
                    try:
                        # Try to parse the date in "месяц.день" format
                        dt = datetime.strptime(date, '%b.%d')
                    except ValueError:
                        try:
                            # Try to parse the date in "год" format
                            dt = datetime.strptime(date, '%Y')
                        except ValueError:
                            return ''
        # return dt.strptime('%d.%m.%Y')
        return dt
    
dft['Начало трудового стажа'] = dft['Начало трудового стажа'].apply(lambda x: format_date(x))
dft['Дата рождения'] = dft['Дата рождения'].apply(lambda x: format_date(x))
dft['Начало трудовой деятельности в РОСАТОМ'] = dft['Начало трудовой деятельности в РОСАТОМ'].apply(lambda x: format_date(x))
```

## Возраст, стаж

```python
def count_years_since(date):
    now = datetime.now()
    delta = now - date
    years = delta.days / 365.25
    return years
```

```python
dft['Возраст'] = dft['Дата рождения'].apply(lambda x: count_years_since(x))
dft['Стаж'] = dft['Начало трудового стажа'].apply(lambda x: count_years_since(x))
dft['Стаж РОСАТОМ'] = dft['Начало трудовой деятельности в РОСАТОМ'].apply(lambda x: count_years_since(x))

dft.head(10)
```

## Знак зодиака

```python
def get_zodiac_sign(dat):
    day = dat.day
    month = dat.month
    if (month == 1 and day >= 20) or (month == 2 and day <= 18):
        return 'Водолей'
    elif (month == 2 and day >= 19) or (month == 3 and day <= 20):
        return 'Рыбы'
    elif (month == 3 and day >= 21) or (month == 4 and day <= 19):
        return 'Овен'
    elif (month == 4 and day >= 20) or (month == 5 and day <= 20):
        return 'Телец'
    elif (month == 5 and day >= 21) or (month == 6 and day <= 20):
        return 'Близнецы'
    elif (month == 6 and day >= 21) or (month == 7 and day <= 22):
        return 'Рак'
    elif (month == 7 and day >= 23) or (month == 8 and day <= 22):
        return 'Лев'
    elif (month == 8 and day >= 23) or (month == 9 and day <= 22):
        return 'Дева'
    elif (month == 9 and day >= 23) or (month == 10 and day <= 22):
        return 'Весы'
    elif (month == 10 and day >= 23) or (month == 11 and day <= 21):
        return 'Скорпион'
    elif (month == 11 and day >= 22) or (month == 12 and day <= 21):
        return 'Стрелец'
    else:
        return 'Козерог'
    
d = datetime.now()
get_zodiac_sign(d)
```

```python
dft['Знак зодиака'] = dft['Дата рождения'].apply(lambda x: get_zodiac_sign(x))
```
