```python
import pandas as pd
from datetime import datetime
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='most_frequent')
n=3
def format_date(date):
    if pd.isnull(date):  
        return ''
    else:
        try:
            # Try to parse the date in "YYYY-MM-DD" format
            dt = datetime.strptime(date, '%Y-%m-%d')
        except ValueError:
            try:
                # Try to parse the date in "DD.MM.YYYY" format
                dt = datetime.strptime(date, '%d.%m.%Y')
            except ValueError:
                try:
                    # Try to parse the date in "месяц.год" format
                    dt = datetime.strptime(date, '%b.%Y')
                except ValueError:
                    try:
                        # Try to parse the date in "месяц.день" format
                        dt = datetime.strptime(date, '%b.%d')
                    except ValueError:
                        try:
                            # Try to parse the date in "год" format
                            dt = datetime.strptime(date, '%Y')
                        except ValueError:
                            return ''
        return dt.strftime('%d.%m.%Y')

df = pd.read_csv('scores.csv')

df['Начало трудового стажа'] = df['Начало трудового стажа'].apply(format_date)
df['Образование'] = df['Образование'].str.lower()
tmp = pd.DataFrame()
df['Образование'] = df['Образование'].apply(lambda x: 'высшее' if 'высшее' in str(x) else ('среднее' if 'среднее' in str(x) else 'нету'))
encoder = LabelEncoder()

df['Начало трудового стажа'] =  pd.to_datetime(df['Начало трудового стажа'].apply(format_date))
tmp['Начало трудового стажа'] = (df['Начало трудового стажа'] - datetime(1970, 1, 1)).dt.total_seconds()
tmp['Начало трудового стажа']=imputer.fit_transform(tmp['Начало трудового стажа'].values.reshape(-1, 1))
tmp['Образование']=encoder.fit_transform(df['Образование'])
tmp['Образование']=imputer.fit_transform(tmp['Образование'].values.reshape(-1, 1))


features = tmp[['Образование', 'Начало трудового стажа']]  
kmeans = KMeans(n_clusters=n)  
kmeans.fit(features)
predicted_clusters = kmeans.predict(features)
df['Cluster'] = predicted_clusters
cluster_means = df.groupby('Cluster')['Total.real.%'].mean()
print(cluster_means)

for cluster_id in range(n):  
    cluster_data = df[df['Cluster'] == cluster_id]
    fios = cluster_data['ФИО'].tolist()  
    scores = cluster_data['Total.real.%'].tolist() 
    print(f"Кластер {cluster_id}:")
    for fio, score in zip(fios, scores):
        print(f"ФИО: {fio}, Баллы: {score}")
    print()
```
